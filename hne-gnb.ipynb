A Pipeline is built using RandomUnderSampler and SMOTE from the 'imblearn' package to randomly undersample the majority class and the create synthetuc data for the minority class simultaneously.

```python
xvals_train_filtered = np.log(xvals_train_filtered)
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
over = SMOTE(sampling_strategy=0.2)
under = RandomUnderSampler(sampling_strategy=0.8)
from imblearn.pipeline import Pipeline
steps = [('o', over), ('u', under)]
pipeline = Pipeline(steps=steps)
X_over, y_over = pipeline.fit_resample(xvals_train_filtered, yvals_train_binary)
```

Gaussian Naive Bayes classifier is used, since the input variables are continuous. The input variables have to follow a normal distribution, so we take the log of the input values. 

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_over, y_over)
xvals_test_filtered = np.log(xvals_test_filtered)
testX_over, testY_over = pipeline.fit_resample(xvals_test_filtered, yvals_test_binary)
```

The summary of the Gaussian Naive Bayes classifier is as shown below.

```python
from sklearn.metrics import classification_report, confusion_matrix
y_pred = nb.predict(testX_over)
print(classification_report(testY_over, y_pred))
confusion_matrix(testY_over, nb.predict(testX_over))
```

Precision-recall curve is plotted to evaluate performance of model.

```python
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import f1_score
from sklearn.metrics import auc
# predict probabilities
lr_probs = nb.predict_proba(testX_over)
# keep probabilities for the positive outcome only
lr_probs = lr_probs[:, 1]
# predict class values
yhat = nb.predict(testX_over)
lr_precision, lr_recall, _ = precision_recall_curve(testY_over, lr_probs)
lr_f1, lr_auc = f1_score(testY_over, yhat), auc(lr_recall, lr_precision)
# summarize scores
print('GNB: f1=%.3f auc=%.3f' % (lr_f1, lr_auc))
# plot the precision-recall curves
no_skill = len(testY_over[testY_over==1]) / len(testY_over)
plt.plot([0, 1], [no_skill, no_skill], linestyle='--', label='No Skill')
plt.plot(lr_recall, lr_precision, marker='.', label='GNB')
# axis labels
plt.xlabel('Recall')
plt.ylabel('Precision')
# show the legend
plt.legend()
# show the plot
plt.show()
```
